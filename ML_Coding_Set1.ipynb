```python
#1.
import pandas as pd
data = {'Customer_ID': [101, 102, 103, 101, 104, 102, 101, 105, 102, 103],
        'Purchase_Amount': [200, 150, 180, 220, 300, 200, 100, 400, 250, 300]}
data = pd.DataFrame(data)
Total_pur = data.groupby('Customer_ID').sum().reset_index()
print("Total Purchases per Customer:\n",Total_pur)
freq_cust = Total_pur.sort_values('Purchase_Amount',ascending=False).head(3)
print("Top 3 Frequent Customers:\n",freq_cust)
```

```python
#2.
import pandas as pd
data = {'Square_Feet': [1500, 2000, 2500, 3000, 3500],
        'Price': [300000, 400000, 500000, 600000, 700000]}
data = pd.DataFrame(data)
independent = data[['Square_Feet']]
dependent =  data[['Price']]
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(independent,dependent)
Test_data = [[1800],[2800]]
predict = regressor.predict(Test_data)
print("Predicted Prices:\n",predict)
```

```python
#3.
import pandas as pd
data = {'Category': ['A', 'B', 'C', 'A', 'D', 'B', 'A', 'E', 'B', 'C', 'C', 'A'],
        'Value': [10, 15, 20, 30, 25, 18, 22, 40, 35, 50, 45, 15]}
data = pd.DataFrame(data)
grouped_sum = data.groupby('Category').size().reset_index(name='Count')
print("Total Occurrences per Category:\n",grouped_sum)
frequent = grouped_sum.sort_values('Count',ascending=False).head(3)
print("Top 3 Frequent Categories:\n",frequent)
```

```python
#4.
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, None, 35, None]}

dataset = pd.DataFrame(data)
print("Original Data:")
print(dataset)

dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
print("Data after Imputation:")
print(dataset)
```

```python
#5.
import pandas as pd
import numpy 

data = {'Experience': [1, 2, 3, 4, 5], 
        'Salary': [30000, 35000, 40000, 45000, 50000]}
df = pd.DataFrame(data)

x_train =df[['Experience']]
y_train =df[['Salary']]

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)

x_test = [[6]]

predict_sal = regressor.predict(x_test)
print("Predicted Salary for 6 years of experience:$",predict_sal[0])
```

```python
from sklearn.preprocessing import LabelEncoder
data = {'word_count': [100, 150, 200, 120, 180, 220], 
        'is_spam': ['ham', 'spam', 'spam', 'ham', 'spam', 'spam']}
df = pd.DataFrame(data)

x_train = df[['word_count']]
encoder = LabelEncoder()
y_train_encoded= encoder.fit_transform(df[['is_spam']])

from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(x_train,y_train_encoded)

x_test = [[120]]

prediction = regressor.predict(x_test)

pre_output = encoder.inverse_transform(prediction)

print("Prediction for email with 200 words:",pre_output[0])
```
